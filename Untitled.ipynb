```python
flower = imread("flower.jpg")
flower = resize(flower, (360,360,3))
plt.imshow(flower)
plt.show()
```

```python
def image_to_pandas(image):
    df = pd.DataFrame([image[:,:,0].flatten(), image[:,:,1].flatten(), image[:,:,2].flatten()]).T
    df.columns = ["Red_Channel", "Green_Channel", "Blue_Channel"]
    return df
```

```python
def plot_pixels(df):
    x_ = df["Red_Channel"]
    y_ = df["Green_Channel"]
    z_ = df["Blue_Channel"]
    
    color_list = list(zip(
                        list(df['Red_Channel']),
                        list(df["Green_Channel"]),
                        list(df["Blue_Channel"])))
    fig = plt.figure(figsize = (12,10))
    ax_3d = plt.axes(projection = "3d")
    ax_3d.scatter3D(xs = x_, ys = y_, zs = z_, alpha = 0.55)
    ax_3d.set_xlim3d(0, x_.max())
    ax_3d.set_ylim3d(0, y_.max())
    ax_3d.set_zlim3d(0, z_.max())
    ax_3d.invert_yaxis()
```
